### 1. What is data ingestion and why is it important in data science? Which Python library is commonly used for data ingestion and cleaning?

**Data Ingestion** refers to the process of transferring data from one or more sources into a system where it can be stored, processed, and analyzed. This can involve data from various formats such as databases, files (CSV, JSON, XML), APIs, streaming data, etc.

**Importance in Data Science:**
*   **Foundation for Analysis:** Before any analysis or machine learning can begin, the data must first be brought into a usable environment. Ingestion is the crucial first step.
*   **Access to Diverse Sources:** Data scientists often need to work with data scattered across many systems. Ingestion tools allow for consolidating this data.
*   **Real-time Insights:** For applications requiring immediate decisions (e.g., fraud detection, personalized recommendations), real-time data ingestion is vital.
*   **Scalability:** As data volumes grow, efficient ingestion processes are necessary to handle the load.

**Common Python library for data ingestion and cleaning:**

**Pandas** is by far the most commonly used Python library for data ingestion (especially from structured files like CSV, Excel, SQL databases) and cleaning. Its DataFrame structure provides powerful and flexible tools for reading, manipulating, and cleaning tabular data.

### 2. Why is data cleaning necessary before performing analysis or machine learning?

Data cleaning (also known as data scrubbing or data cleansing) is a critical preprocessing step that involves detecting and correcting or removing corrupt or inaccurate records from a record set, table, or database. It is essential before performing analysis or machine learning for several reasons:

*   **Ensures Data Quality:** Dirty data can lead to inaccurate conclusions and poor model performance. Cleaning improves the overall quality, consistency, and reliability of the data.
*   **Improves Model Accuracy:** Machine learning models are highly sensitive to the quality of the input data. Inconsistent formats, outliers, or missing values can significantly degrade model accuracy and lead to biased results.
*   **Prevents Misleading Insights:** In data analysis, uncleaned data can result in incorrect statistical summaries, misleading trends, and flawed visualizations, leading to poor business decisions.
*   **Handles Missing Values:** Missing data is a common issue that can cause errors or make certain analyses impossible. Cleaning involves strategies like imputation or removal of rows/columns to address this.
*   **Corrects Inconsistent Formats:** Data from different sources often have varying formats (e.g., dates, text strings). Cleaning standardizes these formats, making the data uniform and easier to process.
*   **Removes Duplicates:** Duplicate records can skew analyses and inflate the importance of certain observations. Cleaning identifies and removes these redundant entries.
*   **Manages Outliers:** Extreme values (outliers) can disproportionately influence statistical models. Cleaning involves identifying and deciding how to handle these outliers (e.g., removal, transformation).
*   **Enhances Data Usability:** Clean data is easier to understand, work with, and integrate into various tools and workflows, saving time and effort in subsequent steps.

### 4. What is the difference between `head()` and `tail()` functions in Pandas?

Both `head()` and `tail()` are convenient Pandas DataFrame methods used to quickly inspect the contents of a DataFrame, but they display different parts of the DataFrame:

*   **`head(n=5)`**: This method returns the **first `n` rows** of the DataFrame. By default, if no argument is passed, it returns the first 5 rows. It's useful for a quick look at the beginning of the data, especially to check column names, data types, and initial data patterns.

*   **`tail(n=5)`**: This method returns the **last `n` rows** of the DataFrame. By default, if no argument is passed, it returns the last 5 rows. It's useful for checking the end of the data, which can be important for time-series data or to see how data ends after certain operations.

**Example:**

```python
import pandas as pd

# Create a sample DataFrame
data = {'A': range(1, 11), 'B': [f'val_{i}' for i in range(1, 11)]}
df_sample = pd.DataFrame(data)

print("Original DataFrame (df_sample):")
display(df_sample)

print("\nUsing head(3) to see the first 3 rows:")
display(df_sample.head(3))

print("\nUsing tail(2) to see the last 2 rows:")
display(df_sample.tail(2))
```

### 5. How can you identify missing values in a dataset?

Identifying missing values is a crucial step in data cleaning and preprocessing. Missing data can arise for various reasons, such as data entry errors, data corruption, or simply not being recorded. Pandas provides several powerful and flexible methods to detect and quantify missing values.

Missing values are typically represented as `NaN` (Not a Number) for numerical data, or sometimes `None` for object-type data in Python, which Pandas often converts to `NaN` internally. Other representations like `?`, `NA`, or empty strings might also represent missing data and need to be handled during data ingestion or cleaning.

Here are the common methods to identify missing values in a Pandas DataFrame:

1.  **`df.isnull()` or `df.isna()`:**
    These methods return a boolean DataFrame of the same shape as the original DataFrame, where `True` indicates a missing value and `False` indicates a non-missing value. `isnull()` and `isna()` are aliases for each other.

    *   **Usage:** `df.isnull()`

2.  **`df.isnull().sum()`:**
    To get a count of missing values per column, you can chain the `sum()` method after `isnull()`. Since `True` is treated as 1 and `False` as 0, `sum()` will count the `True` values (i.e., missing values) in each column.

    *   **Usage:** `df.isnull().sum()`
    *   **Output:** A Series where the index is the column name and the value is the total count of missing values in that column.

3.  **`df.isnull().sum().sum()`:**
    To get the total count of missing values across the entire DataFrame, you can chain `sum()` twice. The first `sum()` gives counts per column, and the second `sum()` adds these up.

    *   **Usage:** `df.isnull().sum().sum()`
    *   **Output:** A single integer representing the total number of missing values.

4.  **`df.isnull().any()`:**
    This method returns a boolean Series indicating whether *any* missing values exist in each column.

    *   **Usage:** `df.isnull().any()`
    *   **Output:** A Series where the index is the column name and the value is `True` if the column contains at least one missing value, `False` otherwise.

5.  **`df.isnull().sum() / len(df)` or `df.isnull().mean()`:**
    To get the percentage of missing values per column, you can divide the sum of nulls by the total number of rows (`len(df)`), or simply use `df.isnull().mean()`. This is very useful for deciding whether to drop a column or impute values, especially if a column has a very high percentage of missing data.

    *   **Usage:** `df.isnull().mean() * 100` (for percentage)

6.  **Visualizing Missing Data (using libraries like Missingno):**
    For larger datasets, visualizing missing data can provide quick insights into patterns of missingness. Libraries like `missingno` (which is not part of default Pandas but easily installable) offer functions like `msno.matrix()`, `msno.bar()`, and `msno.heatmap()` to graphically represent missing data.

    *   **`msno.matrix()`:** Shows a nullity matrix, where white lines indicate missing values.
    *   **`msno.bar()`:** A bar chart showing the completeness of each column.
    *   **`msno.heatmap()`:** Shows the correlation of missingness between columns.

```python
import pandas as pd
import numpy as np
# import missingno as msno # Uncomment and install if you want to use for visualization

# Create a sample DataFrame with missing values
data = {
    'A': [1, 2, np.nan, 4, 5, np.nan, 7, 8, 9, 10],
    'B': ['X', 'Y', 'Z', np.nan, 'W', 'V', 'U', np.nan, 'T', 'S'],
    'C': [True, False, True, False, True, False, np.nan, False, True, False],
    'D': [100, 200, 300, 400, 500, 600, 700, 800, 900, np.nan]
}
df_missing = pd.DataFrame(data)

print("Original DataFrame with missing values:")
display(df_missing)

print("\n1. Boolean DataFrame indicating missing values (df.isnull()):")
display(df_missing.isnull())

print("\n2. Count of missing values per column (df.isnull().sum()):")
display(df_missing.isnull().sum())

print("\n3. Total count of missing values in the DataFrame (df.isnull().sum().sum()):")
display(df_missing.isnull().sum().sum())

print("\n4. Check if any missing values exist per column (df.isnull().any()):")
display(df_missing.isnull().any())

print("\n5. Percentage of missing values per column (df.isnull().mean() * 100):")
display(df_missing.isnull().mean() * 100)

# If you have missingno installed, you can visualize:
# print("\nMissing values visualization (Missingno matrix):")
# msno.matrix(df_missing)
# plt.show()

# print("\nMissing values visualization (Missingno bar plot):")
# msno.bar(df_missing)
# plt.show()
```

```python
import pandas as pd
import numpy as np
import io

print("--- Data Ingestion and Cleaning Tasks ---")

# --- Common Setup: Create a sample DataFrame for demonstration ---
# This DataFrame will be used to demonstrate various cleaning techniques.
csv_data = io.StringIO("""
ID,Name,Age,Salary,Department,Experience,City,Status
1,Alice,25,50000,HR,2,New York,Active
2,Bob,30,60000,IT,5,London,Active
3,Charlie,,75000,HR,8,Paris,Inactive
4,David,40,90000,IT,12,New York,Active
5,Eve,22,45000,HR,1,London,Active
6,Frank,35,,Sales,7,Paris,Active
7,Grace,28,55000,IT,3,New York,Inactive
8,Heidi,25,50000,HR,2,New York,Active
9,Ivan,50,120000,,15,London,Active
10,Judy,32,62000,IT,6,,Inactive
11,Karl,29,58000,Sales,4,Paris,Active
12,Liam,38,80000,HR,10,New York,Active
13,Mia,26,52000,IT,3,London,Active
14,Nora,45,,Sales,18,Paris,Active
15,Oscar,31,68000,IT,7,New York,Inactive
""")
df = pd.read_csv(csv_data)

print("\nInitial DataFrame:")
display(df.head())
print(f"Initial DataFrame shape: {df.shape}")

# --- Weak Students (Basic) ---
print("\n--- Weak Students (Basic Tasks) ---")

# 1. Load a CSV file using pandas and display first five rows.
# (Already done above for setup, demonstrating again explicitly)
print("\n1. Loaded CSV file and displayed first five rows:")
# For this example, we use the 'df' created above
display(df.head())

# 2. Identify missing values in a dataset.
print("\n2. Identifying missing values (count per column):")
missing_values_count = df.isnull().sum()
display(missing_values_count)

# 3. Replace missing values with zero.
# Note: This is a basic strategy, often not ideal for all data types.
# We'll create a copy to not affect subsequent tasks immediately.
df_zero_filled = df.fillna(0)
print("\n3. DataFrame after replacing missing values with 0:")
display(df_zero_filled.head())

# 4. Rename column names to lowercase.
df.columns = df.columns.str.lower()
print("\n4. DataFrame with column names converted to lowercase:")
display(df.head())

# 5. Display dataset shape (rows and columns).
print("\n5. Dataset shape (rows, columns):", df.shape)

# --- Normal Students (Moderate) ---
print("\n--- Normal Students (Moderate Tasks) ---")

# 6. Remove duplicate rows from a dataset.
# Create a duplicate for demonstration if not already present
df_with_duplicates = pd.concat([df, df.iloc[[0, 7]]], ignore_index=True)
print("\nDataFrame with artificial duplicates for demonstration:")
display(df_with_duplicates.tail())
print(f"Shape with duplicates: {df_with_duplicates.shape}")

df_no_duplicates = df_with_duplicates.drop_duplicates()
print("\n6. DataFrame after removing duplicate rows:")
display(df_no_duplicates.tail())
print(f"Shape after removing duplicates: {df_no_duplicates.shape}")

# Use the original df (with lowercase columns) for further tasks
df_processed = df.copy()

# 7. Handle missing values using mean or median.
# For 'age' (numerical), use median as it's less sensitive to outliers.
# For 'salary' (numerical), use mean.
# For 'department' (categorical), use mode.
# For 'experience' (numerical), use median.
# For 'city' (categorical), use mode.

# Age: Fill with median
df_processed['age'].fillna(df_processed['age'].median(), inplace=True)

# Salary: Fill with mean
df_processed['salary'].fillna(df_processed['salary'].mean(), inplace=True)

# Department: Fill with mode
# Check for NaN in 'department' after initial fill (if any were present)
if df_processed['department'].isnull().any():
    mode_department = df_processed['department'].mode()[0]
    df_processed['department'].fillna(mode_department, inplace=True)

# Experience: Fill with median (if there were missing values)
df_processed['experience'].fillna(df_processed['experience'].median(), inplace=True)

# City: Fill with mode
if df_processed['city'].isnull().any():
    mode_city = df_processed['city'].mode()[0]
    df_processed['city'].fillna(mode_city, inplace=True)

print("\n7. DataFrame after handling missing values with mean/median/mode:")
display(df_processed.head(10))
print("Missing values after handling:")
display(df_processed.isnull().sum())

# 8. Change data types of selected columns.
# Example: Convert 'experience' to integer type if it became float due to NaN handling
# First, ensure all values are non-null floats to convert to int cleanly
df_processed['experience'] = df_processed['experience'].astype(int)
print("\n8. Data types after changing 'experience' to integer:")
display(df_processed.dtypes)

# 9. Filter rows based on a given condition.
# Example: Employees with salary > 60000 and department 'IT'
filtered_df = df_processed[(df_processed['salary'] > 60000) & (df_processed['department'] == 'IT')]
print("\n9. Filtered DataFrame (Salary > 60000 and Department = 'IT'):")
display(filtered_df)

# 10. Save the cleaned dataset to a new CSV file.
output_csv_path = 'cleaned_data.csv'
df_processed.to_csv(output_csv_path, index=False)
print(f"\n10. Cleaned DataFrame saved to '{output_csv_path}'")
# To verify, you can load it back:
# loaded_df = pd.read_csv(output_csv_path)
# print("Loaded cleaned_data.csv:")
# display(loaded_df.head())

# --- Bright Students (Advanced) ---
print("\n--- Bright Students (Advanced Tasks) ---")

# 11. Compare different missing value handling strategies.
print("\n11. Comparing different missing value handling strategies:")
print("   - **Dropping rows (`df.dropna()`):** Simple, but can lead to significant data loss if many rows have missing values.")
print("     Example: `df.dropna(subset=['age', 'salary'])` removes rows where 'age' or 'salary' is missing.")
print("   - **Imputation (mean/median/mode):** Fills missing values with a calculated statistic. Good for numerical data; mode for categorical. Can reduce variance and introduce bias if not careful.")
print("     Example: `df['age'].fillna(df['age'].median())` (demonstrated in task 7).")
print("   - **Forward/Backward Fill (`df.ffill()`, `df.bfill()`):** Useful for time-series data or when sequential order matters. Fills with the next or previous valid observation.")
print("     Example: `df['salary'].ffill()` fills missing salaries with the previous non-missing salary.")
print("   - **Advanced Imputation (e.g., K-Nearest Neighbors, Regression):** Uses more sophisticated models to predict missing values based on other features. More accurate but computationally intensive.")

# 12. Clean a dataset containing inconsistent categorical values.
# Let's introduce some inconsistencies in 'department' and 'city' for demonstration
df_inconsistent = df_processed.copy()
df_inconsistent.loc[0, 'department'] = ' hR '
df_inconsistent.loc[1, 'department'] = 'it'
df_inconsistent.loc[2, 'city'] = 'paris'
df_inconsistent.loc[9, 'city'] = ' '

print("\nOriginal inconsistent categorical values in 'department':", df_inconsistent['department'].unique())
print("Original inconsistent categorical values in 'city':", df_inconsistent['city'].unique())

df_inconsistent['department'] = df_inconsistent['department'].str.strip().str.capitalize()
df_inconsistent['city'] = df_inconsistent['city'].str.strip().replace('', np.nan).fillna(df_inconsistent['city'].mode()[0]) # Handle empty strings and fill with mode

print("\n12. Cleaned categorical values in 'department':", df_inconsistent['department'].unique())
print("Cleaned categorical values in 'city':", df_inconsistent['city'].unique())
print("DataFrame head after categorical cleaning:")
display(df_inconsistent.head())

# 13. Automate data cleaning steps using functions.
print("\n13. Automating data cleaning steps with a function:")
def clean_data(dataframe):
    df_cleaned = dataframe.copy()

    # 1. Lowercase column names
    df_cleaned.columns = df_cleaned.columns.str.lower()

    # 2. Remove duplicates
    df_cleaned.drop_duplicates(inplace=True)

    # 3. Handle missing numerical values (mean/median)
    numerical_cols = ['age', 'salary', 'experience']
    for col in numerical_cols:
        if col in df_cleaned.columns and df_cleaned[col].isnull().any():
            df_cleaned[col].fillna(df_cleaned[col].median(), inplace=True)

    # 4. Handle missing categorical values (mode) and inconsistencies
    categorical_cols = ['department', 'city', 'status']
    for col in categorical_cols:
        if col in df_cleaned.columns:
            # Clean inconsistencies first (strip whitespace, normalize case)
            df_cleaned[col] = df_cleaned[col].astype(str).str.strip().str.capitalize()
            # Replace empty strings (if any after strip) with NaN for fillna
            df_cleaned[col].replace('', np.nan, inplace=True)
            if df_cleaned[col].isnull().any():
                mode_val = df_cleaned[col].mode()[0]
                df_cleaned[col].fillna(mode_val, inplace=True)

    # 5. Convert 'experience' to integer type if it's float
    if 'experience' in df_cleaned.columns:
        df_cleaned['experience'] = df_cleaned['experience'].astype(int)

    return df_cleaned

# Let's create a fresh messy DataFrame to test the function
csv_messy_data = io.StringIO("""
ID, NAME, AGE, SALARY, DEPARTMENT, Experience, City, Status
1,Alice,25,50000,HR,2,New York,Active
2,Bob,30,,IT,5,London,Active
3,Charlie,,75000,hr,8,Paris,Inactive
4,David,40,90000,IT,12,NEW YORK,Active
5,Eve,22,45000,HR,1,London,Active
6,Frank,35,,Sales,7,Paris,Active
7,Grace,28,55000,IT,3,New York,Inactive
1,Alice,25,50000,HR,2,New York,Active
8,Heidi,25,50000,HR,2,New York,Active
9,Ivan,50,120000,,15,,Active
10,Judy,32,62000,IT,6,London,Inactive
10,Judy,32,62000,IT,6,London,Inactive
""")
df_messy = pd.read_csv(csv_messy_data)
print("Messy DataFrame before cleaning function:")
display(df_messy)

df_auto_cleaned = clean_data(df_messy)
print("\nDataFrame after automated cleaning function:")
display(df_auto_cleaned)
print("Missing values after automated cleaning:")
display(df_auto_cleaned.isnull().sum())

# 14. Perform outlier detection during data cleaning (using IQR method for 'salary').
print("\n14. Outlier detection for 'salary' using IQR method:")
Q1 = df_auto_cleaned['salary'].quantile(0.25)
Q3 = df_auto_cleaned['salary'].quantile(0.75)
IQR = Q3 - Q1
upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR

outliers_salary = df_auto_cleaned[(df_auto_cleaned['salary'] < lower_bound) | (df_auto_cleaned['salary'] > upper_bound)]

print(f"Q1 (Salary): {Q1}, Q3 (Salary): {Q3}, IQR: {IQR}")
print(f"Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")

if not outliers_salary.empty:
    print("Outliers in 'salary' column:")
    display(outliers_salary)
    # Example of handling: remove outliers (for demonstration)
    df_no_outliers = df_auto_cleaned[~((df_auto_cleaned['salary'] < lower_bound) | (df_auto_cleaned['salary'] > upper_bound))]
    print("DataFrame after removing salary outliers:")
    display(df_no_outliers.head())
    print(f"Shape before outlier removal: {df_auto_cleaned.shape}, Shape after: {df_no_outliers.shape}")
else:
    print("No significant outliers found in 'salary' using IQR method in this sample.")
    df_no_outliers = df_auto_cleaned.copy()


# 15. Generate a data cleaning report summarizing actions taken.
print("\n--- 15. Data Cleaning Report ---")
report = []
report.append("Data Cleaning Report Summary:")
report.append("------------------------------")
report.append("\n1. Initial Data Inspection:")
report.append(f"   - Initial DataFrame Shape: {df.shape}")
report.append(f"   - Initial Missing Values:\n{missing_values_count.to_string()}")

report.append("\n2. Basic Cleaning Steps:")
report.append("   - Column names converted to lowercase.")

report.append("\n3. Moderate Cleaning Steps:")
report.append(f"   - {len(df_with_duplicates) - len(df_no_duplicates)} duplicate rows removed.")
report.append("   - Missing 'age' values imputed with median.")
report.append("   - Missing 'salary' values imputed with mean.")
report.append("   - Missing 'department' values imputed with mode.")
report.append("   - Missing 'city' values imputed with mode.")
report.append("   - 'experience' column data type changed to integer.")
report.append("   - Filtered rows based on 'salary' > 60000 and 'department' == 'IT'.")
report.append(f"   - Cleaned DataFrame saved to '{output_csv_path}'.")

report.append("\n4. Advanced Cleaning Steps:")
report.append("   - Discussed different missing value handling strategies (drop, impute, ffill/bfill, advanced).")
report.append("   - Cleaned inconsistent categorical values (whitespace, case normalization) in 'department' and 'city'.")
report.append("   - Implemented an automated `clean_data` function encompassing multiple cleaning steps.")
if not outliers_salary.empty:
    report.append(f"   - Identified and removed {len(outliers_salary)} outliers in 'salary' column using IQR method (bounds: [{lower_bound:.2f}, {upper_bound:.2f}]).")
else:
    report.append("   - Performed outlier detection for 'salary' using IQR; no significant outliers found.")

report.append(f"\nFinal Cleaned DataFrame Shape (after outlier removal): {df_no_outliers.shape}")

print('\n'.join(report))
print("\n--- End of Data Ingestion and Cleaning Tasks ---")
```
