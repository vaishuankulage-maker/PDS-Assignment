### 1. Difference between Binomial, Poisson, and Normal distributions

Probability distributions are mathematical functions that describe the likelihood of different outcomes in a random process. Here's how Binomial, Poisson, and Normal distributions differ:

*   **Binomial Distribution:**
    *   **Nature:** Discrete probability distribution.
    *   **Scenario:** Models the number of successes in a fixed number of independent Bernoulli trials (experiments with only two outcomes: success or failure).
    *   **Key Characteristics:**
        *   **Fixed Number of Trials (n):** The experiment is repeated a set number of times.
        *   **Two Possible Outcomes:** Each trial results in either a 'success' or a 'failure'.
        *   **Independent Trials:** The outcome of one trial does not affect the outcome of another.
        *   **Constant Probability of Success (p):** The probability of success remains the same for every trial.
    *   **Parameters:** `n` (number of trials) and `p` (probability of success).
    *   **Examples:** The number of heads in 10 coin flips, the number of defective items in a sample of 50 products.

*   **Poisson Distribution:**
    *   **Nature:** Discrete probability distribution.
    *   **Scenario:** Models the number of times an event occurs in a fixed interval of time or space, given that these events occur with a known constant mean rate and independently of the time since the last event.
    *   **Key Characteristics:**
        *   **Events are Rare:** Typically used for events that occur infrequently.
        *   **Events are Independent:** The occurrence of one event does not affect the probability of another.
        *   **Constant Rate (λ):** Events occur at a constant average rate over the interval.
        *   **No Upper Limit:** The number of occurrences is theoretically unlimited within the interval.
    *   **Parameter:** `λ` (lambda), which represents the average rate of event occurrence.
    *   **Examples:** The number of phone calls received by a call center in an hour, the number of typos on a page of a book, the number of customers arriving at a shop in a specific time frame.

*   **Normal (Gaussian) Distribution:**
    *   **Nature:** Continuous probability distribution.
    *   **Scenario:** Describes data that clusters around a central value with fewer values further away. It's often called the 'bell curve' due to its characteristic symmetric shape.

### 3. What is the Central Limit Theorem and how does it relate to Normal distribution?

The **Central Limit Theorem (CLT)** is one of the most fundamental and powerful theorems in statistics. It states that:

> *Regardless of the original distribution of the population*, the sampling distribution of the sample mean (or sum) of a sufficiently large number of independent, identically distributed random variables will be approximately normally distributed.

**Key Aspects of CLT:**
*   **Sample Size:** The approximation to normality improves as the sample size (`n`) increases. A common rule of thumb is that `n ≥ 30` is generally sufficient, though it can be smaller for populations that are already somewhat symmetric.
*   **Independence:** The samples must be independent of each other.
*   **Identically Distributed:** The random variables must be drawn from the same population distribution.

**Relation to Normal Distribution:**
The CLT explains *why* the Normal distribution is so ubiquitous in statistics, even when individual data points don't appear to be normally distributed. Here's how it relates:

1.  **Foundation for Statistical Inference:** The CLT is the cornerstone of many statistical hypothesis tests and confidence intervals. When we take repeated samples from a population and calculate their means, the distribution of these sample means will tend to be normal. This allows us to use normal distribution theory to make inferences about the population mean, even if we don't know the population's original distribution.

2.  **Ubiquity of Normality:** Many phenomena in nature and social sciences involve the aggregation or sum of many small, independent random effects. For example, a person's height is influenced by many genetic and environmental factors. The CLT suggests that the sum of these many independent influences would result in a normal distribution of heights in a population.

3.  **Approximation for Other Distributions:** For large `n`, the binomial and Poisson distributions can be approximated by the normal distribution. For a Binomial distribution, when `np > 5` and `n(1-p) > 5`, it can be approximated by a normal distribution with `μ = np` and `σ² = np(1-p)`. For a Poisson distribution, when `λ > 10`, it can be approximated by a normal distribution with `μ = λ` and `σ² = λ`.

In essence, the CLT provides a powerful theoretical justification for using normal distribution models for sample statistics (like the mean), making it indispensable for drawing conclusions about populations based on sample data.

### 5. What is skewness in distributions and which distributions exhibit it?

**Skewness** is a measure of the asymmetry of the probability distribution of a real-valued random variable about its mean. In simpler terms, it tells us about the shape of the distribution, specifically if it's lopsided to one side.

*   **Symmetric Distribution (Zero Skewness):** If a distribution is perfectly symmetrical, its skewness is zero. The normal distribution is a classic example of a symmetric distribution, where the mean, median, and mode are all equal and located at the center. The data is evenly distributed on both sides of the mean.

*   **Positive Skewness (Right-Skewed):** A distribution is positively skewed (or right-skewed) if its tail is longer on the right side. This means that the majority of the data points are concentrated on the left side, and there are some larger values that stretch the tail to the right. In a positively skewed distribution, the mean is typically greater than the median, which is greater than the mode (Mean > Median > Mode). This is because the outliers on the right pull the mean in that direction.
    *   **Examples:** Income distribution (a few high earners pull the mean up), time taken to complete a task (most people finish quickly, but a few take much longer), number of children per household.

*   **Negative Skewness (Left-Skewed):** A distribution is negatively skewed (or left-skewed) if its tail is longer on the left side. This indicates that the majority of the data points are concentrated on the right side, and there are some smaller values that stretch the tail to the left. In a negatively skewed distribution, the mean is typically less than the median, which is less than the mode (Mean < Median < Mode). The outliers on the left pull the mean down.
    *   **Examples:** Test scores (most students score high, but a few perform poorly), age of death (most people live to an old age, but some die young), battery life of a product.

**Which distributions exhibit skewness?**

*   **Binomial Distribution:** Can be skewed. It is symmetric only when the probability of success `p = 0.5`. If `p < 0.5`, it is positively skewed. If `p > 0.5`, it is negatively skewed. As the number of trials `n` increases, the binomial distribution tends towards symmetry and can be approximated by the normal distribution.

*   **Poisson Distribution:** Is typically positively skewed. The degree of positive skewness decreases as the mean (λ) increases. For large λ (e.g., λ > 10), the Poisson distribution becomes more symmetric and can be approximated by the normal distribution.

*   **Normal Distribution:** By definition, the normal distribution is perfectly symmetric and has zero skewness.

*   **Other Distributions that exhibit skewness:**
    *   **Exponential Distribution:** Always positively skewed. It models the time between events in a Poisson process.
    *   **Log-Normal Distribution:** Always positively skewed. It is used to model variables whose logarithm is normally distributed.
    *   **Chi-Squared Distribution:** Always positively skewed. The skewness decreases as the degrees of freedom increase.
    *   **Weibull Distribution:** Can be either positively or negatively skewed, depending on its shape parameter. It's often used in reliability engineering.
    *   **Beta Distribution:** Can be positively, negatively, or even symmetric, depending on its parameters.

```python
import numpy as np
from scipy.stats import binom, poisson, norm
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
```

### 1. Binomial Distribution

#### Theory:

The Binomial distribution models the number of successes in a fixed number of independent Bernoulli trials, each with the same probability of success. It has two parameters: `n` (number of trials) and `p` (probability of success).

*   **Probability Mass Function (PMF):** $P(X=k) = \binom{n}{k} p^k (1-p)^{n-k}$

#### Implementation and Analysis:

```python
# Parameters for Binomial Distribution
n_binom = 10  # Number of trials
p_binom = 0.5 # Probability of success

# Generate a range of possible outcomes (number of successes)
k_binom = np.arange(0, n_binom + 1)

# Calculate the Probability Mass Function (PMF)
pmf_binom = binom.pmf(k_binom, n_binom, p_binom)

# Calculate Mean and Variance
mean_binom = binom.mean(n_binom, p_binom)
var_binom = binom.var(n_binom, p_binom)

print(f"Binomial Distribution (n={n_binom}, p={p_binom}):")
print(f"  Mean: {mean_binom:.2f}")
print(f"  Variance: {var_binom:.2f}")

# Plotting the Binomial Distribution
plt.figure(figsize=(10, 6))
plt.bar(k_binom, pmf_binom, color='skyblue', edgecolor='black')
plt.title(f'Binomial Distribution (n={n_binom}, p={p_binom})')
plt.xlabel('Number of Successes (k)')
plt.ylabel('Probability')
plt.xticks(k_binom)
plt.show()
```

### 2. Poisson Distribution

#### Theory:

The Poisson distribution models the number of times an event occurs in a fixed interval of time or space, given that these events occur with a known constant mean rate (λ) and independently of the time since the last event. It is a discrete distribution.

*   **Probability Mass Function (PMF):** $P(X=k) = \frac{\lambda^k e^{-\lambda}}{k!}$

#### Implementation and Analysis:

```python
# Parameters for Poisson Distribution
lambda_poisson = 3 # Average rate of event occurrence

# Generate a range of possible outcomes (number of events)
k_poisson = np.arange(0, 10) # Adjust upper limit based on lambda

# Calculate the Probability Mass Function (PMF)
pmf_poisson = poisson.pmf(k_poisson, lambda_poisson)

# Calculate Mean and Variance
mean_poisson = poisson.mean(lambda_poisson)
var_poisson = poisson.var(lambda_poisson)

print(f"Poisson Distribution (λ={lambda_poisson}):")
print(f"  Mean: {mean_poisson:.2f}")
print(f"  Variance: {var_poisson:.2f}")

# Plotting the Poisson Distribution
plt.figure(figsize=(10, 6))
plt.bar(k_poisson, pmf_poisson, color='lightcoral', edgecolor='black')
plt.title(f'Poisson Distribution (λ={lambda_poisson})')
plt.xlabel('Number of Events (k)')
plt.ylabel('Probability')
plt.xticks(k_poisson)
plt.show()
```

### 3. Normal (Gaussian) Distribution

#### Theory:

The Normal distribution is a continuous probability distribution that is symmetric and bell-shaped. It is characterized by two parameters: the mean (μ) and the standard deviation (σ). Many natural phenomena follow a normal distribution.

*   **Probability Density Function (PDF):** $f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{1}{2} \left(\frac{x-\mu}{\sigma}\right)^2}$

#### Implementation and Analysis:

```python
# Parameters for Normal Distribution
mu_normal = 0    # Mean
sigma_normal = 1 # Standard Deviation

# Generate a range of values (x-axis for continuous distribution)
x_normal = np.linspace(mu_normal - 4*sigma_normal, mu_normal + 4*sigma_normal, 100)

# Calculate the Probability Density Function (PDF)
pdf_normal = norm.pdf(x_normal, mu_normal, sigma_normal)

# Calculate Mean and Variance
mean_normal = norm.mean(loc=mu_normal, scale=sigma_normal)
var_normal = norm.var(loc=mu_normal, scale=sigma_normal)

print(f"Normal Distribution (μ={mu_normal}, σ={sigma_normal}):")
print(f"  Mean: {mean_normal:.2f}")
print(f"  Variance: {var_normal:.2f}")

# Plotting the Normal Distribution
plt.figure(figsize=(10, 6))
plt.plot(x_normal, pdf_normal, color='mediumseagreen', linewidth=2)
plt.title(f'Normal Distribution (μ={mu_normal}, σ={sigma_normal})')
plt.xlabel('Value (x)')
plt.ylabel('Probability Density')
plt.fill_between(x_normal, pdf_normal, color='mediumseagreen', alpha=0.3)
plt.show()
```
